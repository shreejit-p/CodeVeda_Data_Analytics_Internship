## Task 3: Basic Data Visualization – Iris Dataset 
This notebook creates and customizes basic plots and exports them as images.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Loading the dataset
df = pd.read_csv("cleaned_iris.csv")
df.head()

```

```python
# CREATE AND CUSTOMISE PLOTS
# Bar plot- Species Distribution (Categorical Data)
species_counts = df["species"].value_counts()
species_counts #Calculates how many flower belongs to each species
```

```python
#Bar Plot- Species distribution
plt.figure(figsize=(7,5))

for species in species_counts.index:
    plt.bar([species], [species_counts[species]], label=species)

plt.xlabel("Iris Species")
plt.ylabel("Number of Samples")
plt.title("Distribution of Iris Species")
plt.ylim(0,80)
plt.legend()
plt.savefig("species_distribution_barplot.png", dpi=300, bbox_inches="tight") #Saves the barplot

plt.show()


```

```python
#Line chart- To observe trend of variables
plt.figure(figsize=(8,5))
plt.plot(df["sepal_length"])

plt.xlabel("Sample Index")
plt.ylabel("Sepal Length (cm)")
plt.title("Sepal Length Across All Samples")
plt.savefig("sepal_length_lineplot.png", dpi=300, bbox_inches="tight")# Saves the line plot

plt.show()
```

```python
#Line chart- To observe trend of variables
plt.figure(figsize=(8,5))
plt.plot(df["petal_length"])

plt.xlabel("Sample Index")
plt.ylabel("Petal Length (cm)")
plt.title("Petal Length Across All Samples")
plt.savefig("petal_length_lineplot.png", dpi=300, bbox_inches="tight")# Saves the line plot

plt.show()
```

```python
#In the above line plot we can see three distinct pattern of petal lengths across samples, this supports the species variation of the dataset.
```

```python
#Scatter plot - To Visualise relationships between variables

plt.figure(figsize=(7,5))

for species in df["species"].unique():
    subset = df[df["species"] == species]
    plt.scatter(subset["petal_length"], subset["petal_width"], label=species)

plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.title("Petal Length vs Petal Width by Species")
plt.legend()
plt.savefig("petal_scatter_by_species.png", dpi=300, bbox_inches="tight") #Saves the scatter plot
plt.show()
```
