## Task 2: Exploratory Data Analysis (EDA)  
This notebook explores the cleaned dataset using statistics and visualizations.


```python
import pandas as pd
#Loading the cleaned data set
df = pd.read_csv("cleaned_iris.csv")
df.head()
```

```python
#Getting the summary statistics of the data
df.describe()
df.describe(include="all")
```

#Univariate analysis(One variable at a time)

```python
#A) Histogram- Distribution of sepal length
import matplotlib.pyplot as plt

plt.figure()
plt.hist(df["sepal_length"], bins=10)
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Sepal Length")
plt.show()
```

```python
#Histogram- Distribution of sepal length

plt.figure()
plt.hist(df["petal_length"], bins=10)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length")
plt.show()
```

```python
# B) Boxplot- to detect outliers
#for sepal length
plt.figure()
plt.boxplot(df["sepal_length"])
plt.title("Boxplot of Sepal Length")
plt.show()

#for petal length
plt.figure()
plt.boxplot(df["petal_length"])
plt.title("Boxplot of Petal Length")
plt.show()
```

```python
#The Boxplot of petal length shows that the data is right skewed
```

```python
species_counts = df["species"].value_counts()

plt.figure()
plt.bar(species_counts.index, species_counts.values)
plt.xlabel("Species")
plt.ylabel("Count")
plt.title("Distribution of Iris Species")
plt.show()
```

#Bivariate analysis

```python
#Scatter plot- sepal length vs sepal width

plt.figure()
plt.scatter(df["sepal_length"], df["sepal_width"])
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal Length vs Sepal Width")
plt.show()
```

```python
#Scatter Plot- Petal Length vs petal width
plt.figure()
plt.scatter(df["petal_length"], df["petal_width"])
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal Length vs Petal Width")
plt.show()

```

```python
#Correlation Matrix
corr_matrix = df.corr(numeric_only=True)
corr_matrix
```

```python
# Interpretation of the correlation matric:
#The correlation matrix shows that petal length and petal width are very strongly positively correlated (r â‰ˆ 0.96), indicating they grow together. 
#Sepal length also shows strong positive relationships with both petal length (r â‰ˆ 0.87) and petal width (r â‰ˆ 0.82).
#In contrast, sepal width has weak to moderate negative correlations with petal measurements and almost no correlation with sepal length.
```
